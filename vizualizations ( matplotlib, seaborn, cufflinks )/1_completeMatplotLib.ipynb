```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 5, 45)

y = x**3
print(x, y)
```

```python
#creating a linear graph with x-axis and y-axis using arrays
plt.plot(x, y)

#adds title for the plots
plt.title("Hello World")

#names the x-axis
plt.xlabel("kaka")

#names the y-axis
plt.ylabel("kaki")
```

```python
#subplots in matplotlib
# subpot(rows, cols, selection of plot) takes 3 arguements 

# if we give the rows and cols as 2 then the matplotlib plane divides into 4 sub plots and we can create 4 subplots
# in order to select a particular subplot we can use the following plt.subplot(2, 2, 1) -> 1st sub plot

plt.subplot(2, 2, 1)
plt.subplot(2, 2, 2)
plt.subplot(2, 2, 3)
plt.subplot(2, 2, 4)
```

```python
# in order to execute any graph in any of the subplots we can simply write the graph code below the selected subplot
plt.subplot(2, 2, 1)
plt.subplot(2, 2, 2)
plt.plot(x, y)
plt.subplot(2, 2, 3)
plt.plot(x**2 , x)
plt.subplot(2, 2, 4)
```

```python
# Object Oriented Way
# There is a drawback for subplots that we cannot overlap 2 subplots 
# we use figure object which is the matplotlib canvas 

#fig.add_axis(left spacing, bottom spacing, height, width)

#figsize is a tuple which consists of ( width, height )
# dpi is nothing but number of pixels higher the dpi higher the clarity of the graph

fig = plt.figure(figsize=(10, 6), dpi=800)

axis1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axis1.plot(x, y)

axis2 = fig.add_axes([0.1, 0.1, 0.3, 0.4])
axis2.plot(x**2, x)
```

```python
#Types of plots

#it shows the graphs or the relations among multiple data points as dotted format used for regression analysis
plt.scatter(x, y)
```

```python
#histogram
# histogram only consists of 1 arguement it is mainly used for visualizing the distribution of the data points
from random import sample
data = sample(range(1, 1000), 100)

plt.hist(data)
```

```python
#boxplot 
#it is mainly used for statistical analysis it shows mean median and mode

plt.boxplot(x)

```

```python
#saving plots
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.plot(x, y)

#stores in the same directory
plt.savefig("plot.png")
```

```python
#working with images

import matplotlib.image as mpimg
```

```python
#it converts the readed image in the format of array ( divides the image based on the pixels and converts into ndim arrays )

img = mpimg.imread("image.jpg")
img
```

```python
#to see image
plt.imshow(img)

#to remove axis
plt.axis('off')
```

```python
#crop the image
cropped_img = img[50:100, 100:300]  #crop: rows 50-200, cols 100-300
plt.imshow(cropped_img)
plt.axis('off')
```
