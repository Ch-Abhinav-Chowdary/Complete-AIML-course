# Matrix Plots

Matrix plots are visualization techniques used to analyze relationships, correlations, patterns, and missing values across multiple variables in a dataset. They are especially useful in Exploratory Data Analysis (EDA) because they help identify trends, dependencies, clusters, and data quality issues quickly.

It has mainly 3 types they are

1) heatmap

2) Clustermap

3) pivot tables

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

# Heatmap

A Heatmap is a matrix-based visualization that represents numerical data using different colors to show the magnitude of values. It is mainly used to identify patterns, correlations, trends, and relationships between variables in a dataset. In data analysis, heatmaps are commonly used to display correlation matrices, where each cell represents the correlation value between two variables. Strong positive or negative relationships are highlighted using intense colors, while weak relationships appear lighter. Heatmaps make large amounts of data easier to interpret visually and are widely used in Exploratory Data Analysis (EDA), machine learning, business analytics, and statistical analysis to quickly detect hidden insights and dependencies among features.

```python
flights = sns.load_dataset('flights')
flights
```

```python
tips = sns.load_dataset('tips')
```

Correlation and heatmaps are directly connected because a heatmap is one of the best ways to visually represent correlation values between variables.

Suppose you have a dataset with columns like:

Age

Salary

Experience

Working Hours

When you calculate correlation, you get numerical values showing how strongly each variable is related to another. Instead of reading those numbers manually in a table, a heatmap converts them into colors, making relationships easier to understand visually.

For example:

Correlation Data Table

| Variables | Correlation |
|-----------|-------------|
| Age vs Salary | 0.85 |
| Salary vs Experience | 0.92 |
| Age vs Working Hours | -0.20 |

In a heatmap:

Strong positive correlations (0.85, 0.92) appear with darker or intense colors.
Weak correlations (0.10, -0.20) appear with lighter colors.
Negative correlations often use contrasting colors.

So, a heatmap helps you instantly identify:

Which variables are strongly related

Which variables move together

Which variables move oppositely

Which variables have almost no relationship

In machine learning and data analysis, correlation heatmaps are mainly used for:

Feature selection

Detecting multicollinearity

Understanding data patterns

Removing redundant features


If 2 variables in a correlation has 0 value then they are independent 

if they are closer to -1 then they have inverse relationship
if they are closer to +1 then they have direct relationship

```python
tipscorr = tips[['total_bill', 'tip', 'size']]
tipscorr
```

```python
tipscorr.corr()
```

```python
sns.heatmap(tipscorr.corr(), annot=True)  #annot shows they values inside the boxes 
```

# Clustermaps

A Cluster Map (Clustermap) is an advanced matrix visualization that combines a heatmap with hierarchical clustering to group similar rows and columns together automatically. It not only displays data values using colors like a heatmap but also rearranges the data based on similarity patterns, making hidden relationships easier to identify. The clustering process creates tree-like structures called dendrograms on the rows and columns, which show how closely related different variables or observations are. Clustermaps are widely used in data science, machine learning, bioinformatics, customer segmentation, and pattern recognition to detect groups, trends, and similarities within complex datasets. For example, in a customer dataset, a clustermap can automatically group customers with similar purchasing behavior or group features that behave similarly.

```python
sns.clustermap(tipscorr.corr())
```

# pivot table heatmap

A Pivot Table Heatmap is a visualization technique where a pivot table is combined with a heatmap to summarize and visually analyze relationships between categorical and numerical data. A pivot table first organizes and aggregates data into a matrix format using rows, columns, and calculated values such as sum, average, or count. Then, a heatmap applies colors to the matrix values, where higher values appear with stronger or darker colors and lower values appear lighter. This makes it easier to identify patterns, trends, high-performing categories, and anomalies instantly without manually reading large tables of numbers. Pivot table heatmaps are commonly used in business analytics, sales analysis, financial reporting, and data science for comparing performance across multiple dimensions. For example, a company can create a pivot table heatmap showing total sales across different products and regions, where darker colors immediately highlight regions with the highest sales.

```python
pvtflight = flights.pivot_table(values='passengers', index='month', columns='year')
pvtflight
```

```python
sns.heatmap(pvtflight, annot=True)
```
