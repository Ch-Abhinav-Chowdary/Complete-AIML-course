Distribution plots are used to understand how data values are spread across a dataset. They help you see patterns like:

Where most values are concentrated

Whether data is symmetric or skewed

Presence of outliers

Spread/variance of data

Shape of the data distribution

Lets use some of the distrubution plots which we are going to use

1) distplot, histplot
2) jointplot
3) pairplot
4) rugplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# there are some datasets are already present in seaborn library
df = sns.load_dataset('tips')  # make sure internet connection is stable or not

#manually decide which attribute is qualitative and quantitative
```

```python
df['size'].unique()
```

```python
sns.distplot(df['total_bill'], bins=10)

#it shows the output in the bins format 
# a bin is a range of value and the individual bar size represents number of values that falls into that bins
# for example from bin 10 to 20 we have maximum distrubution
```

```python
sns.histplot(df['total_bill'], bins=100, kde=True)
```

```python
plt.subplot(1, 2, 1)
sns.histplot(df['total_bill'], bins=10, kde=True)
plt.subplot(1, 2, 2)
sns.histplot(df['tip'], bins=10, kde=True)
```

```python
#jointplot used to compare 2 attribute values
sns.jointplot(x='total_bill', y='tip', data=df, kind='scatter')
sns.jointplot(x='total_bill', y='tip', data=df, kind='kde')
sns.jointplot(x='total_bill', y='tip', data=df, kind='hex')
sns.jointplot(x='total_bill', y='tip', data=df, kind='reg')
```

```python
#pairplot it is most important plot in representing distribution data maily used to compare all the attributes which consists of numerical values
sns.pairplot(df, hue='sex')
```

```python
sns.pairplot(df, hue='size', palette='rainbow')
```

```python
#rugplot - it is similar to blanket

sns.rugplot(df['total_bill'])
sns.rugplot(df)
```
