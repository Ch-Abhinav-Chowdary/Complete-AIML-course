# Regression plot

Regression plots are visualization techniques used to analyze the relationship between variables and understand how one variable changes with respect to another. They mainly help identify trends, correlations, and predictive relationships in data. A regression plot usually contains scatter plot points along with a fitted regression line that represents the best possible linear relationship between the variables. The regression line helps estimate how the dependent variable changes when the independent variable changes. These plots are widely used in statistics, machine learning, and data analysis for trend analysis, prediction, and understanding feature relationships. For example, in a dataset containing study hours and exam marks, a regression plot can show whether increasing study time generally leads to higher marks. If the points closely follow the regression line, it indicates a strong relationship, while scattered points suggest a weaker relationship.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

# lmplot()

An LM Plot (lmplot) is a Seaborn visualization used to display the relationship between two numerical variables along with a fitted linear regression line. It combines a scatter plot and a regression line to help analyze trends, correlations, and predictive relationships in data. The scatter points represent actual observations, while the regression line shows the best-fit line that models the relationship between the variables. lmplot is especially useful for identifying positive, negative, or weak correlations and can also separate data using categories with different colors through the hue parameter. It is widely used in Exploratory Data Analysis (EDA), statistics, and machine learning to understand how variables are related and whether a linear pattern exists between them.

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.lmplot(x='total_bill', y='tip', data = tips)
```

```python
sns.lmplot(x='total_bill', y='tip', data = tips, hue='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data = tips, hue='sex', palette='coolwarm')
```

```python
sns.lmplot(x='total_bill', y='tip', data = tips, hue='sex', palette='coolwarm', markers=['o', 'v'])
```

```python
sns.lmplot(x='total_bill', y='tip', data = tips, hue='sex', palette='coolwarm', markers=['o', 'v'], scatter_kws={'s': 1000})
```
