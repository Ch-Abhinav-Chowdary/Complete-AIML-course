Categorical plots are used to visualize relationships between categorical variables and numerical data.

They help answer questions like:

Which category has the highest average?

How are values distributed across groups?

How many observations belong to each category?

Types of categorical plots

1) boxplot
2) violinplot
3) stripplot
4) swarmplot
5) barplot
6) countplot

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df
```

```python
#countplot - shows the count of the categories
sns.countplot(x = df['sex'], hue=df['smoker'])
```

```python
sns.countplot(x = df['sex'])
```

```python
#barplot - it will takes both axis x - categorical data ( qualitative ), y - numerical data ( quantitative )

sns.barplot(x = df['sex'], y = df['total_bill'])  #on default it used average as estimator 
```

```python
import numpy as np

#we can change estimator
sns.barplot(x = df['sex'], y = df['total_bill'], estimator=np.sum)
```

```python
sns.barplot(x = df['sex'], y = df['total_bill'], estimator=np.max)
```

### Boxplot

A Box Plot is a statistical visualization used to summarize the distribution of numerical data using five main values: minimum, first quartile (Q1), median, third quartile (Q3), and maximum. The box represents the interquartile range (IQR), which contains the middle 50% of the data, while the line inside the box shows the median. Whiskers extend to the minimum and maximum values within a specific range, and points outside the whiskers are considered outliers. Box plots are mainly used to compare distributions, detect skewness, and identify outliers efficiently across different categories.

```python
# we will cover it very detail in statistics 
#boxplot

sns.boxplot(x = 'tip', y = 'day', data = df, palette='rainbow')
```

## Vilonplot

A Violin Plot is an advanced distribution plot that combines features of a box plot and a density plot. It not only shows summary statistics such as the median and quartiles but also displays the probability density of the data through its violin-like shape. Wider sections of the violin indicate regions where data values are more concentrated, while narrower sections represent lower density. Violin plots are useful when you want deeper insight into the underlying distribution shape, especially for comparing multiple groups with complex or multi-modal data distributions.

```python
sns.violinplot(x = df['day'], y = df['total_bill'])
```

# stripplot

A Strip Plot is a categorical scatter plot used to display individual data points across different categories. Each point represents a single observation, making it useful for visualizing the exact distribution, spread, and clustering of data values. Since points are plotted directly, overlapping can occur when many values are similar, especially in dense datasets. Strip plots are mainly used for small to medium-sized datasets where seeing every individual observation is important.

```python
sns.stripplot(x = df['sex'], y = df['total_bill'])
```

# swarmplot

A Swarm Plot is an improved version of a strip plot that also displays individual data points but automatically adjusts their positions to prevent overlapping. This creates a more organized and readable visualization while still preserving the exact values of each observation. Swarm plots are particularly useful for dense datasets where strip plots become cluttered, allowing better understanding of concentration, spread, and patterns within categorical groups.

```python
sns.swarmplot(x = df['sex'], y = df['total_bill'])
```

```python
# we can combine both vilon and swarm plots
sns.violinplot(x = 'tip', y = 'day', data = df)
sns.swarmplot(x = 'tip', y = 'day', data = df)
```
