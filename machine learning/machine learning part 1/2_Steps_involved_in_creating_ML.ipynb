# Exploratory Data Analysis (EDA) — clear, practical explanation

What EDA is
- Exploratory Data Analysis (EDA) is the process of investigating and summarizing a dataset before building models. It turns raw data into understanding: patterns, anomalies, relationships, and quality issues. EDA is creative, iterative, and evidence-driven.

Why EDA matters for ML
- Ensures model assumptions are appropriate.
- Reveals data quality problems (missing values, duplicates, wrong types).
- Identifies informative features and useless/noisy ones.
- Guides feature engineering, transformation, and selection.
- Prevents wasted effort on models trained on flawed data.

Main goals
1. Understand data structure and types (numerical, categorical, datetime, text).
2. Assess quality (missingness, duplicates, errors).
3. Summarize distributions and variability.
4. Detect outliers and anomalies.
5. Discover relationships and correlations between variables.
6. Generate hypotheses for modeling and feature creation.

Practical step-by-step EDA workflow
1. Quick overview
    - Look at samples (head), shape, dtypes, basic stats (mean, std, min, max).

2. Summary Statistics
    - mean, median, mode, std, min, max, quartiles - helps understand spread and central tendency

3. Value Count
    - How many unique values in a column ?
    - Great for categorical columns

4. Missing Value Analysis
    - Where are the gaps ? what percent of data is missing ?

5. Visualizatiosns
    - Histograms -> Distribution of the values
    - Boxplots -> outliers and spread
    - Barplots -> comparisons of categories
    - Correlation heatmaps -> linear relationship between numerical features
    - Scatter plot -> bivariate relationships

6. Target Variable Exploration
    - Hoe does your output relate to other variables


Common pitfalls
- Ignoring class imbalance.
- Blindly removing outliers without domain context.
- Relying only on correlations (correlation ≠ causation).
- Overlooking data leakage during feature engineering.

Deliverables from EDA
- Cleaned dataset or cleaning plan.
- Candidate features and transformations.
- Shortlist of modeling strategies and baseline metrics to try.
- Visuals and summary report explaining why choices were made.

Bottom line
- EDA is the essential investigative step that turns data into hypotheses and reliable features. Good EDA shortens model development time, reduces surprises, and improves model validity.

# Importance of EDA

1. You can't clean or preprocess what you don't understand
2. It helps you identify mistakes, biases, or limitations in the data
3. It guides the direction of your data cleaning and feature engineering
4. It gives your audience or stakeholders a story or overview of what the data is telling you

# Data Cleaning

EDA tells you what's wrong. Data Cleaning fixes it. Cleaning is not glamorous, but it's 80% of the work in real-world projects.


1. Handling Missing Values
    - Check which columns have missing values
    - Strategies to handle
    - Drop missing rows/columns 
    - impute with:
        - Mean/Median -> for numerical data
        - Mode -> for categorical data
        - Advanced: Linear regression, KNN or interpolation
2. Remove Duplicates
    - Detect and Drop exact duplicate rows
3. Fix Data Types
    - Convert Wrong Types ( e.g., numbers stored as strings, dates as text )
4. Handling Inconsistent Categories
    - Clean up categorical values such as: 
        - "Male", "male", "Male" -> Should all become "male"
5. Detect and Handle Outliers
    - Use boxplots, IQR or Z-Score
    - Handle by:
        - Removing ( if clearly wrong )
        - Capping
6. Fix Logic or Domain Errors
    - E.g., age = -5 is Invalid or BMI = 150 likely an error
    - Can replace with mean, median or remove

# Data Preprocessing

- To prepare clean data so it can be analyzed or used in machine learning model
- If Data Cleaning is about fixing mistakes then Data Processing about transforming valid data into a usable format

# Steps Involved in Data Preprocessing
1. Encoding Categorical Variables

    - Convert text lablels to numerical value
    - Two methods of encoding:

        1. 1 Label Encoding: Good for ordered categories like "Low", "Medium" and "High"

        1. 2 One-Hot Encoding: For non-ordered categories like region

2. Feature Transformation
    - Used to handle skewed data, like right-skewed or left skewed data

3. Feature Scaling ( Normalization or standardization )
    - Bring numerical values to the same scale especially useful for distance-based algorithms

    - Types of Feature Scaling

        3. 1 Normalization (Min-Max Scaling): scales values between 0 and 1

        3. 2 Standardization (Z-Score Scaling): Transformation data to have mean 0 and std 1

# Feature Selection

Selecting the most useful features and removing the rest

## Why it is important ?

- Reduce noise and overfitting
- Speeds up training
- Imporves model accuracy
- Makes model Interpretation easier

##### Methods in Feature Selection: 
    
1. Filter Method: 
    - COrrelation matrix -> Removes highly correlated features 
    - Chi-Square test ( Categorical vs Categorical )
    - ANOVA F-test ( numerical vs categorical target )

2. Embedding Methods: 
    - Lasso Regression -> Shrinks coefficients to 0
    - Tree-based models ( Random forest, XGBoost )
        - Feature importance score
