# KNN (K-Nearest Neighbors)

The laziest, yet sometimes most effective machine learning algorithm ever invented. It literally asks: "Who are your friends?" and classifies you based on that.


## 1) Definition of KNN: "Show Me Your Friends and I'll Tell You Who You Are"

**K-Nearest Neighbors (KNN)** is a simple classification algorithm that classifies a new data point based on the class of its K nearest neighbors in the training data.

**The Core Idea:**
- To predict the class of a new point, look at the K closest points in your training data.
- Count how many belong to each class.
- The new point belongs to the class with the most votes (majority voting).

**Why is it called "lazy learning"?**
- KNN does NO training! It literally memorizes the training data.
- All the computation happens during prediction (opposite of eager learners like logistic regression).
- It's like procrastinating till the last minute and then asking your friends for homework help. The actual "studying" happens when you need the answer!

**Key Properties:**
- Non-parametric: doesn't assume any particular distribution of the data.
- Instance-based: makes decisions based on similar examples.
- Memory-intensive: stores the entire training set.
- Sensitive to irrelevant features and data scale.

## 2) How to Determine K: The Goldilocks Problem

Choosing K is like choosing how many friends to ask before making a decision:
- K=1: Ask only your BEST friend (easily influenced, noisy).
- K=5: Ask your close friend group (balanced judgment).
- K=100: Ask everyone in the city (takes forever, likely wrong).

**The K Selection Problem:**

**K too small (K=1):**
- ✓ Fast decisions
- ✗ Very noisy and sensitive to outliers
- ✗ Overfits to training data
- Real-world analogy: Asking only your weirdest friend for advice!

**K too large (K=n):**
- ✓ Smoother decision boundaries
- ✓ Less sensitive to noise
- ✗ Slow predictions
- ✗ Underfits (misses patterns)
- Real-world analogy: Taking a poll of everyone in your city; the majority opinion drowns out local perspectives!

**How to Choose K:**

1. **Rule of thumb:** K = √n, where n = total training samples
2. **Cross-validation:** Try different K values and pick the one with best validation accuracy
3. **Domain knowledge:** Sometimes you know how many neighbors should matter
4. **Odd numbers:** Choose odd K for binary classification (avoids ties)

**General Guidelines:**
- Small datasets: K = 3-5
- Medium datasets: K = 5-15
- Large datasets: K = √n or higher

## 3) The KNN Algorithm: Step by Step

**The Process:**

```
1. Store all training examples (x, y) in memory
2. For each new query point Q:
   a. Calculate distance from Q to all training points
   b. Sort by distance (ascending)
   c. Select the K nearest points
   d. Count votes from each class
   e. Predict the class with the most votes
   f. If regression: average the K nearest values
```

**Distance Metrics:**

The most common is **Euclidean Distance** (straight-line distance):

`d(A, B) = √[(x_A - x_B)² + (y_A - y_B)² + ... + (z_A - z_B)²]`

**Alternative metrics:**
- **Manhattan Distance** (taxi route): `|x_A - x_B| + |y_A - y_B| + ...`
- **Minkowski Distance** (generalization): `(|x_A - x_B|^p + |y_A - y_B|^p + ...)^(1/p)`
- **Hamming Distance** (for categorical data): count differing attributes

**Pseudocode:**
```
function KNN_Classify(new_point, training_data, K):
    distances = []
    for each training_point in training_data:
        dist = calculate_distance(new_point, training_point)
        distances.append((dist, training_point.class))
    
    distances.sort_by_distance()
    k_nearest = distances[:K]
    
    vote_counts = count_votes(k_nearest)
    predicted_class = vote_counts.max_class()
    
    return predicted_class
```

## 4) KNN for Continuous Data (Regression)

**Classification:** Majority vote
**Regression:** Average the K nearest values!

**Example:**
- Predict house price based on similar houses
- Look at K=5 nearest houses (by features)
- Average their prices
- That's your prediction!

**KNN Regression Formula:**
`ŷ = (1/K) * Σ(y_i) for i in K nearest neighbors`

**When to use KNN regression:**
- ✓ Non-linear relationships
- ✓ Local patterns matter
- ✗ Slow for large datasets
- ✗ Hard with many irrelevant features

**Real example:**
```
Training data:
House 1: 1000 sqft, location features X → Sold for $150,000
House 2: 1100 sqft, location features Y → Sold for $165,000
House 3: 950 sqft, location features Z → Sold for $148,000

New house: 1050 sqft, similar location
K=3: Average = (150,000 + 165,000 + 148,000) / 3 = $154,333
```


## 5) Use Cases and Real-World Examples

**When KNN Shines:**

1. **Recommendation Systems**
   - Netflix: "People who watched Game of Thrones also watched..."
   - Spotify: Music recommendations based on similar listeners
   - Amazon: "Customers who bought this also bought..."

2. **Credit Scoring & Loan Approval**
   - Look at K similar applicants
   - If most got approved, approve the new applicant
   - Interpretable: can explain the decision by showing similar cases

3. **Disease Diagnosis**
   - Patient profile vs K similar patient profiles
   - Doctors can understand: "Your case is similar to these 5 patients who had X disease"

4. **Image Classification (with pixel features)**
   - Find K most similar images
   - Vote on the category

5. **Handwriting Recognition**
   - Digit recognition: compare new digit to K nearest training digits
   - Used before neural networks!

6. **Fraud Detection**
   - Transaction vs K similar transactions
   - If K transactions were fraudulent, flag the new one

**When KNN Fails Miserably:**

- ✗ Large datasets (slow predictions)
- ✗ Many irrelevant features (noise dominates)
- ✗ High dimensions (curse of dimensionality)
- ✗ Imbalanced classes (majority always wins)
- ✗ Sensitive to feature scaling


## 6) Worked Example: Predicting Movie Genre with KNN

**Problem:** Predict if a new movie is "Action" or "Romance" based on its features.

**Features:** (Explosions per minute, Love scenes per minute)

**Training Data:**
```
Movie 1: Explosions=8, Love=1 → Action
Movie 2: Explosions=7, Love=2 → Action
Movie 3: Explosions=2, Love=8 → Romance
Movie 4: Explosions=1, Love=9 → Romance
Movie 5: Explosions=6, Love=3 → Action
Movie 6: Explosions=3, Love=7 → Romance
```

**New Movie:** Explosions=5, Love=4 (Unknown genre)

**Using K=3:**

---

**Step 1: Calculate distances from new movie to all training movies**

Using Euclidean distance formula: `d = √[(x1-x2)² + (y1-y2)²]`

Distance to Movie 1:
`d1 = √[(5-8)² + (4-1)²] = √[9 + 9] = √18 ≈ 4.24`

Distance to Movie 2:
`d2 = √[(5-7)² + (4-2)²] = √[4 + 4] = √8 ≈ 2.83` ← NEAREST!

Distance to Movie 3:
`d3 = √[(5-2)² + (4-8)²] = √[9 + 16] = √25 = 5.00`

Distance to Movie 4:
`d4 = √[(5-1)² + (4-9)²] = √[16 + 25] = √41 ≈ 6.40` ← FARTHEST!

Distance to Movie 5:
`d5 = √[(5-6)² + (4-3)²] = √[1 + 1] = √2 ≈ 1.41` ← VERY NEAREST!

Distance to Movie 6:
`d6 = √[(5-3)² + (4-7)²] = √[4 + 9] = √13 ≈ 3.61`

---

**Step 2: Sort by distance and select K=3 nearest**

Sorted:
```
Movie 5: d=1.41 → Action
Movie 2: d=2.83 → Action
Movie 6: d=3.61 → Romance
```

**Step 3: Count votes**
```
Action votes: 2
Romance votes: 1
```

**Step 4: Make prediction**

**PREDICTION: ACTION**

Why? Because 2 out of 3 nearest neighbors are Action movies. The new movie (5 explosions, 4 love scenes) is more similar to action movies!

---

**Visualization of the Decision:**
```
                    New Movie (5,4) ★
                    /  \  \
                 2.83  1.41  3.61
                  /      |      \
        Movie2(7,2)   Movie5(6,3)  Movie6(3,7)
          Action        Action      Romance
          
Majority: 2 Action vs 1 Romance → PREDICT ACTION
```

---

**What if we used K=5?**

The 5 nearest:
```
Movie 5: Action (1.41)
Movie 2: Action (2.83)
Movie 6: Romance (3.61)
Movie 1: Action (4.24)
Movie 3: Romance (5.00)
```

Votes: 3 Action, 2 Romance → Still PREDICT ACTION!

---

**What if we used K=1?**

Only Movie 5 (closest): Action → PREDICT ACTION

(Same result, but more risky! If Movie 5 was mislabeled, we'd be fooled.)

---

**Sensitivity Analysis:**

If features were scaled differently:
- Without scaling: feature with larger range dominates
- With scaling: features weighted equally

**Example without scaling:**
```
Features: (Reviews count: 0-10000, Rating: 0-5)
New data: (5000 reviews, 4 rating) vs Training data
→ Reviews count overwhelms rating!

Solution: Normalize both to [0,1]:
Reviews: 5000/10000 = 0.5
Rating: 4/5 = 0.8
→ Both features now equally important
```

This is why **Feature Scaling** is CRITICAL for KNN!


## Summary: KNN in Nutshell

**Pros:**
- ✓ Simple to understand and implement
- ✓ Non-parametric (no assumptions about data distribution)
- ✓ Effective for local patterns
- ✓ Works for both classification and regression
- ✓ Can handle non-linear boundaries

**Cons:**
- ✗ Slow predictions (compute distance to ALL training points)
- ✗ Requires feature scaling
- ✗ Sensitive to irrelevant features (curse of dimensionality)
- ✗ Stores entire training set (memory-intensive)
- ✗ Majority voting with imbalanced classes

**The Bottom Line:**

KNN is the "ask your friends" algorithm. It doesn't learn patterns; it memorizes the data and predicts based on similarity. It's perfect for small datasets but struggles with big data.

Think of it as the social network effect: *"You are the average of your 5 nearest neighbors."*
