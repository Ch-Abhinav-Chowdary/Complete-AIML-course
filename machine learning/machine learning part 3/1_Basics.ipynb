# What is Classification ?

Classification is a supervised learning task where the goal is to assign each input to one of several discrete categories. The output is a label, not a number.

- Examples: cat vs dog, spam vs not spam, disease vs healthy.
- The model learns from labeled examples and then predicts the class for new data.


## Key Concepts in Classification

- **Classes / labels**: the categories the model predicts.
- **Features**: the input values used to make predictions.
- **Training set**: labeled examples used to teach the model.
- **Prediction**: the class the model assigns to a new example.
- **Decision boundary**: the line or surface separating classes in feature space.


## How Classification Works

The model uses training examples to learn the relationship between features and labels. For binary classification, it typically computes a score and compares it to a threshold.

In many classifiers, the decision boundary is the location where the predicted score changes class. For example, logistic regression uses the sigmoid function to convert a linear score into a probability, then applies a threshold at 0.5.

Logistic regression prediction formula:

`σ(z) = 1 / (1 + e^{-z})`, where `z = w₀ + w₁ x₁ + w₂ x₂ + ...`

If `σ(z) ≥ 0.5`, predict class 1; otherwise predict class 0.

## Real-Time Examples

- **Email spam detection**: classify messages as spam or not spam.
- **Medical diagnosis**: predict whether a patient has a disease or is healthy.
- **Loan approval**: decide whether a borrower is likely to repay a loan.
- **Image recognition**: identify handwritten digits or object categories.


## Model Evaluation

Classification models are evaluated using a confusion matrix, which summarizes true and false predictions.

- **True Positive (TP)**: correctly predicted positive examples.
- **False Positive (FP)**: negative examples incorrectly predicted positive.
- **True Negative (TN)**: correctly predicted negative examples.
- **False Negative (FN)**: positive examples incorrectly predicted negative.


![Confusion Matrix](images/classification_confusion_matrix.png)

This confusion matrix shows how many examples were classified correctly and incorrectly. It is the basis for accuracy, precision, recall, and F1 score.


## Derived Metrics

- **Accuracy**: overall fraction of correct predictions.
  `Accuracy = (TP + TN) / (TP + TN + FP + FN)`
- **Precision**: fraction of positive predictions that are correct.
  `Precision = TP / (TP + FP)`
- **Recall**: fraction of actual positives detected.
  `Recall = TP / (TP + FN)`
- **F1 score**: harmonic mean of precision and recall.
  `F1 = 2 * (Precision * Recall) / (Precision + Recall)`


## Example Calculation

Suppose a disease test produces the following counts:
- TP = 50, FP = 10, FN = 5, TN = 35

Then:
- Accuracy = (50 + 35) / (50 + 10 + 5 + 35) = 85 / 100 = 0.85
- Precision = 50 / (50 + 10) = 0.83
- Recall = 50 / (50 + 5) = 0.91
- F1 score = 2 * (0.83 * 0.91) / (0.83 + 0.91) ≈ 0.87

This shows how the confusion matrix values are used to derive each metric.


## Summary

- Classification predicts discrete labels, not continuous values.
- The decision boundary separates classes in feature space.
- The confusion matrix is the foundation for evaluation metrics.
- Precision, recall, and F1 help measure performance when classes are imbalanced.
- Real-time examples include spam detection, medical diagnosis, and loan approval.
