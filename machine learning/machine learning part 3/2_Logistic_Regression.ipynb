# Logistic Regression

A comprehensive guide to understanding logistic regression with clear explanations, derivations, real-time examples, and visual illustrations.

## 1) The Problem with Linear Regression for Classification

Imagine you're working at a bank and want to predict whether a customer will get a loan approved (Yes=1) or rejected (No=0) based on their credit score.

**Why Linear Regression FAILS:**
- Linear regression predicts ANY number: -5, 0.7, 150, etc.
- But we need ONLY 0 or 1 (or at least probabilities between 0 and 1)
- A customer with credit score 100 might predict loan approval as 1.5 (which makes no sense!)
- A customer with credit score -50 might predict -0.8 (negative probability?!)

**The satire:** It's like asking a yes/no question but getting an answer of "42" or "maybe pizza tomorrow." Linear regression is like a confused butler who doesn't understand the difference between a binary choice and a continuous number line!

![Why Linear Regression Fails](images/logistic_linear_regression_fail.png)

**Image explanation:** Notice how the green line (linear fit) goes below 0 on the left and above 1 on the right. Those predictions are IMPOSSIBLE for binary classification! Logistic regression squashes all predictions into the beautiful 0-1 range like a mathematical bodyguard at a club.

## 2) Definition of Logistic Regression

Logistic Regression is a classification algorithm that models the **probability** that an example belongs to class 1 (positive class).

**Key ideas:**
- It outputs a probability between 0 and 1.
- If probability >= 0.5, predict class 1.
- If probability < 0.5, predict class 0.
- It's called "regression" because internally it still uses linear regression, but wraps it with a special function called the sigmoid.

**In simple terms:** Logistic regression answers the question "What's the probability this belongs to class 1?" instead of "Is this class 1 or 0?" The probability makes the difference.

## 3) The Sigmoid Function: The Star of the Show

The sigmoid function is the hero that transforms linear regression into logistic regression. It's defined as:

`σ(z) = 1 / (1 + e^(-z))`

**What it does:**
- Takes ANY number (from -∞ to +∞) as input.
- Outputs a number between 0 and 1.
- When z = 0, σ(z) = 0.5 (the decision boundary).
- As z increases, σ(z) approaches 1.
- As z decreases, σ(z) approaches 0.

**The math behind it:**
- The exponential function e^(-z) is always positive.
- Adding 1 to it keeps it positive.
- Dividing 1 by this always gives a value between 0 and 1.
- It's smooth and differentiable (great for optimization!)

Think of it as a magic squasher that takes any number and squeezes it into the [0, 1] range like a mathematical accordion!

![The Sigmoid Function](images/logistic_sigmoid_function.png)

**Image explanation:** 
- The S-curve shape is the sigmoid function.
- It smoothly transitions from 0 (left) to 1 (right).
- At z=0, the probability is exactly 0.5 (perfect uncertainty).

- This curve is the mathematical foundation of logistic regression!

## 4) The Process of Logistic Regression

**Step 1: Start with a linear combination**
`z = w₀ + w₁x₁ + w₂x₂ + ... + wₙxₙ`

This is just like linear regression. We combine features with weights.

**Step 2: Apply the sigmoid function**
`h(x) = σ(z) = 1 / (1 + e^(-z))`

Now h(x) is a probability between 0 and 1. This is our predicted probability of class 1.

**Step 3: Make a prediction**
- If h(x) >= 0.5, predict class 1.
- If h(x) < 0.5, predict class 0.

**Step 4: Measure the error**
Instead of squared error, we use **cross-entropy loss** (also called log loss):

`J(w) = -(1/m) * Σ[yᵢ * log(hᵢ) + (1 - yᵢ) * log(1 - hᵢ)]`

This penalizes confident wrong predictions heavily!

**Step 5: Optimize the weights**
Use gradient descent to minimize the loss function:

`wⱼ := wⱼ - α * (∂J/∂wⱼ)`

Where α is the learning rate.

The gradient (derivative) of cross-entropy loss is beautifully simple:

`∂J/∂wⱼ = (1/m) * Σ(hᵢ - yᵢ) * xᵢⱼ`

That's the same as linear regression! But applied to probabilities instead of raw values.

## 5) The Intuition Behind Logistic Regression

**Analogy: A Jury Making a Verdict**

Think of logistic regression like a jury deciding whether to convict (class 1) or acquit (class 0):

- **Evidence (Features):** The jury considers different pieces of evidence (x₁, x₂, ..., xₙ).
- **Weights (Learned importance):** Each piece of evidence has a weight (w₁, w₂, ..., wₙ). Some evidence matters more than others!
- **Linear combination:** The jury mentally sums up all the weighted evidence (z = w₀ + w₁x₁ + ...).
- **Sigmoid (Confidence level):** The jury's confidence in conviction is represented as σ(z), a probability between 0 and 1.
- **Decision threshold:** If confidence >= 0.5 (majority of jurors), they convict. Otherwise, they acquit.

**Why it works:**
- Simple to understand: it's just weighted sum + probability.
- Interpretable: you can see which features matter more (larger weights = more important).
- Mathematically elegant: the sigmoid naturally converts any input to a valid probability.

## 6) Worked Example with Step-by-Step Calculations

**Problem:** Predict if a student will pass an exam (Pass=1, Fail=0) based on study hours.

**Training Data:**
- Student 1: 2 hours studied → Failed (0)
- Student 2: 3 hours studied → Failed (0)
- Student 3: 4 hours studied → Passed (1)
- Student 4: 5 hours studied → Passed (1)

**Goal:** Find weights w₀ (bias) and w₁ (coefficient for hours).

---

**Step 1: Initialize weights**
Let's start with: w₀ = 0, w₁ = 0.5 (arbitrary starting point)

---

**Step 2: Calculate predictions for each training example**

**For Student 1 (x₁ = 2, y₁ = 0):**
- z₁ = 0 + 0.5 * 2 = 1.0
- h₁ = σ(1.0) = 1 / (1 + e^(-1.0)) = 1 / (1 + 0.368) = 1 / 1.368 ≈ 0.731
- Predicted probability of passing: **73.1%** (but actually failed, so this is WRONG)

**For Student 2 (x₂ = 3, y₂ = 0):**
- z₂ = 0 + 0.5 * 3 = 1.5
- h₂ = σ(1.5) = 1 / (1 + e^(-1.5)) = 1 / (1 + 0.223) ≈ 0.818
- Predicted probability: **81.8%** (but failed, VERY WRONG!)

**For Student 3 (x₃ = 4, y₃ = 1):**
- z₃ = 0 + 0.5 * 4 = 2.0
- h₃ = σ(2.0) = 1 / (1 + e^(-2.0)) = 1 / (1 + 0.135) ≈ 0.881
- Predicted probability: **88.1%** (actually passed, pretty good!)

**For Student 4 (x₄ = 5, y₄ = 1):**
- z₄ = 0 + 0.5 * 5 = 2.5
- h₄ = σ(2.5) = 1 / (1 + e^(-2.5)) = 1 / (1 + 0.082) ≈ 0.924
- Predicted probability: **92.4%** (actually passed, excellent!)

---

**Step 3: Calculate the cross-entropy loss**

For each example: Loss = -[yᵢ * log(hᵢ) + (1 - yᵢ) * log(1 - hᵢ)]

**Student 1:** Loss₁ = -[0 * log(0.731) + 1 * log(1 - 0.731)] = -log(0.269) ≈ 1.312
**Student 2:** Loss₂ = -[0 * log(0.818) + 1 * log(1 - 0.818)] = -log(0.182) ≈ 1.704
**Student 3:** Loss₃ = -[1 * log(0.881) + 0 * log(1 - 0.881)] = -log(0.881) ≈ 0.127
**Student 4:** Loss₄ = -[1 * log(0.924) + 0 * log(1 - 0.924)] = -log(0.924) ≈ 0.079

**Total Loss:** (1.312 + 1.704 + 0.127 + 0.079) / 4 ≈ **0.805**

---

**Step 4: Update weights using gradient descent**

Gradient for w₁ = (1/4) * Σ(hᵢ - yᵢ) * xᵢ
= (1/4) * [(0.731 - 0) * 2 + (0.818 - 0) * 3 + (0.881 - 1) * 4 + (0.924 - 1) * 5]
= (1/4) * [1.462 + 2.454 - 0.476 - 0.38]
= (1/4) * [3.06]
= 0.765

With learning rate α = 0.1:
**w₁ := w₁ - α * gradient = 0.5 - 0.1 * 0.765 = 0.5 - 0.0765 ≈ 0.423**

Similarly for w₀ (bias), the gradient is:
Gradient for w₀ = (1/4) * Σ(hᵢ - yᵢ)
= (1/4) * [0.731 + 0.818 - 0.119 - 0.076]
= (1/4) * [1.354]
= 0.339

**w₀ := 0 - 0.1 * 0.339 ≈ -0.034**

---

**Step 5: Repeat!**

We repeat steps 2-4 many times (iterations) until the loss stops decreasing significantly. After many iterations, the weights converge and we have our trained model!

**Final trained weights (after convergence):** w₀ ≈ -3.5, w₁ ≈ 0.9

**Making predictions with the trained model:**
- New student studies 4.5 hours: z = -3.5 + 0.9 * 4.5 = 0.55
- h = σ(0.55) ≈ 0.634
- Prediction: **63.4% chance of passing**, so PREDICT PASS (class 1)

---

**Interpretation:**
- The model learned that more study hours increase the probability of passing.
- w₁ = 0.9 means each additional hour roughly increases the log-odds by 0.9.
- w₀ = -3.5 is the baseline (intercept) that shifts the curve left/right.

## Summary: The Big Picture

**Logistic Regression at a glance:**

1. **Input:** Features (x₁, x₂, ..., xₙ)
2. **Linear combination:** z = w₀ + w₁x₁ + w₂x₂ + ...
3. **Sigmoid transformation:** h(x) = 1 / (1 + e^(-z))
4. **Output:** Probability of class 1 (between 0 and 1)
5. **Decision rule:** If h(x) >= 0.5, predict class 1; else class 0
6. **Loss function:** Cross-entropy loss
7. **Optimization:** Gradient descent to minimize loss

**Key takeaways:**
- Solves the problem of linear regression: outputs valid probabilities.
- The sigmoid is the magic ingredient that makes it work.
- Interpretable: weights show feature importance.
- Foundation for neural networks and other advanced algorithms.
- Simple, elegant, and surprisingly powerful!
