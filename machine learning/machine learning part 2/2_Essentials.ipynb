# Essentials

## 1) Train/Test Split

In machine learning, we split our data into two main parts:
- **Training data**: used to teach the model.
- **Testing data**: used to evaluate how well the model works on new examples.

Why this matters:
- A model that performs well on training data may not do well on new data.
- The test set gives a more realistic measure of model performance.

Typical splits are 70/30 or 80/20 for training/testing.

---

## 2) Mean Squared Error (MSE)

Mean Squared Error is a common metric for regression models. It measures how far the predicted values are from the actual values.

Formula:

`MSE = (1/n) * Σ(yᵢ - ŷᵢ)²`

Where:
- `n` is the number of examples.
- `yᵢ` is the actual value.
- `ŷᵢ` is the predicted value.

Why MSE is useful:
- Squaring errors makes sure negative and positive differences do not cancel out.
- Larger errors are penalized more strongly.

---

## 3) Model Evaluation Techniques

Model evaluation tells us how good a regression model is. Important ideas include:

- **Training score**: how well the model fits the training data.
- **Testing score**: how well the model predicts new data.
- **Generalization**: the model's ability to perform well on unseen data.

A good model has:
- low error on the training set,
- low error on the test set,
- and no large gap between them.

---

## 4) Performance Metrics: R² and Adjusted R²

### R² (R-squared)

R² measures how much of the variation in the target value is explained by the model.

`R² = 1 - (SSR / SST)`

Where:
- `SSR` is the sum of squared residuals (errors).
- `SST` is the total sum of squares.

Interpretation:
- `R² = 1` means perfect prediction.
- `R² = 0` means the model does no better than predicting the average.

### Adjusted R²

Adjusted R² adjusts R² for the number of features used.

Why it matters:
- Adding more features can increase R² even if they are not useful.
- Adjusted R² penalizes unnecessary features, helping prevent overfitting.

Adjusted R² is especially important when comparing models with different numbers of features.

---

## 5) Overfitting, Underfitting, and Generalization

### Overfitting

Overfitting happens when a model learns the training data too well, including noise and small fluctuations.

Signs of overfitting:
- Very low error on training data.
- Much higher error on testing data.

### Underfitting

Underfitting happens when the model is too simple to capture the underlying pattern.

Signs of underfitting:
- High error on both training and testing data.
- The model misses important relationships in the data.

### Generalized Model

A generalized model performs well on both training and unseen data.

We want a model that:
- captures the true pattern in the data,
- avoids noise,
- and predicts accurately on new examples.
