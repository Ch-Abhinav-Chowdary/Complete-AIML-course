# What is Regression ?

Regression is a type of supervised learning used to predict continuous numeric values. In regression, the model learns from examples where the input features and the output numbers are both known. Once trained, the model can estimate values for new data.

Examples of regression tasks:
- Predicting a house price from its size and location.
- Estimating sales based on advertising spend.
- Forecasting temperature based on weather conditions.

## Regression vs Classification

- Regression predicts a continuous quantity, such as price, height, or salary.
- Classification predicts a category, such as "spam vs not spam" or "cat vs dog."

This notebook focuses on regression, which means we are working with numbers that can take any value within a range.

## How Regression Works

The regression model learns a mathematical relationship between input features and the output value. For example, if we want to predict house price from house size, the model finds how the price changes when the size changes.

A simple regression model looks like a straight line in a plot. That is why this method is called **linear regression**.

# Linear Regression

Linear regression is the simplest regression algorithm. It fits a straight line through the training data so the predictions are as close as possible to the actual target values.

The basic formula is:

`y = mx + b`

Where:
- `x` is the input feature.
- `y` is the predicted output.
- `m` is the slope of the line.
- `b` is the intercept, the value of `y` when `x` is 0.

## Linear Regression in Machine Learning Notation

Machine learning uses the hypothesis function:

`h_θ(x) = θ₀ + θ₁ x`

Here:
- `h_θ(x)` is the predicted value for input `x`.
- `θ₀` is the intercept (similar to `b`).
- `θ₁` is the slope of the line (similar to `m`).

The model learns the best values for `θ₀` and `θ₁` from the training data.

## Cost Function and Error

To measure how well the model predicts, we use a cost function. For linear regression, a common cost is the mean squared error:

`J(θ₀, θ₁) = (1 / (2m)) * Σ(h_θ(xᵢ) - yᵢ)²`

Where:
- `m` is the number of training examples.
- `xᵢ` is the input for example `i`.
- `yᵢ` is the actual value for example `i`.
- `h_θ(xᵢ)` is the predicted value.

The cost tells us how far the predictions are from the actual values. The goal is to minimize this cost.

## Gradient Descent

Gradient descent is an algorithm used to find the best line by minimizing the cost function. It updates the parameters `θ₀` and `θ₁` step by step.

The update rules are:

`θ₀ := θ₀ - α * (1/m) * Σ(h_θ(xᵢ) - yᵢ)`
`θ₁ := θ₁ - α * (1/m) * Σ((h_θ(xᵢ) - yᵢ) * xᵢ)`

Where `α` is the learning rate. A smaller `α` means smaller steps toward the best fit.

## What is Global Minima?

The global minima is the point where the cost function is smallest. For linear regression, it means the line fits the data as well as possible.

The training process stops when the cost no longer decreases or when the model reaches the best parameter values.

## Real-Time Example: Predicting House Price

Imagine you want to predict the selling price of a house based on its size in square feet.

Training data might look like this:
- House of 1,000 sq ft sells for $150,000
- House of 1,200 sq ft sells for $180,000
- House of 1,500 sq ft sells for $220,000

Linear regression finds the best line that matches these points. Once trained, the model can predict the price of a new house from its size.

For example, if the model learns that price increases by $120 per square foot and the intercept is $30,000, the prediction is:

`predicted price = 30,000 + 120 * size`

So a 1,350 sq ft house would be predicted as:

`30,000 + 120 * 1350 = $192,000`

## Why Linear Regression is Useful

- It is easy to understand and visualize.
- It works well when the relationship between input and output is roughly linear.
- It is the first step before moving to more complex regression methods.

However, if the data is not linear, other regression models may be needed.

## Summary

- Regression predicts continuous numeric values.
- Linear regression fits a straight line to the data.
- The model uses parameters `θ₀` and `θ₁` to make predictions.
- Cost functions and gradient descent help the model learn the best parameters.
- Real-time examples include house pricing, sales forecasting, and temperature prediction.
